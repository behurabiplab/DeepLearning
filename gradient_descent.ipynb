```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp 
```

```python
x = sp.Symbol('x') #declaring symbolic variable
f = 4*x**2 + 5*x + 8 
print(f"f={f}") #symbolic function
f_derivative_1st = sp.diff( f, x)
print(f"Derivative of function f is {f_derivative_1st}")

#converting symbolic function to numpy numerical compatible function
f = sp.lambdify(x,f,'numpy')
f_derivative_1st = sp.lambdify(x,f_derivative_1st,'numpy')

# Evaluate the function and its derivative over the range of x
x = np.linspace(-3,3,2000)
f_numpy = f(x)
f_derivative_1st_numpy = f_derivative_1st(x) 
print(f"type of f {type(f)}")
```

```python
plt.plot(x,f_numpy, label = "f(x)")
plt.plot(x,f_derivative_1st_numpy, label ="f'(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.title('Function and its Derivative')
plt.legend()
plt.grid()
plt.show()
```

## Finding the Local Minimum of \( f(x) = 4x^2 + 5x + 8 \)

### Step 1: Find the First Derivative
The first derivative of \( f(x) \) is:

$$ f'(x) = \frac{d}{dx} (4x^2 + 5x + 8) = 8x + 5 $$

### Step 2: Set the First Derivative to Zero
To find the critical points, we set the first derivative equal to zero:

$$ 8x + 5 = 0 $$

Solving for \( x \):

$$ x = -\frac{5}{8} $$

### Step 3: Find the Second Derivative
To determine if this is a minimum or maximum, we compute the second derivative:

$$ f''(x) = \frac{d}{dx} (8x + 5) = 8 $$

Since \( f''(x) = 8 > 0 \), the function is **concave up**, confirming that \( x = -\frac{5}{8} \) is a **local minimum**.

### Step 4: Compute the Minimum Value
To find the minimum value, substitute \( x = -\frac{5}{8} \) into \( f(x) \):

$$ f\left(-\frac{5}{8}\right) = 4\left(-\frac{5}{8}\right)^2 + 5\left(-\frac{5}{8}\right) + 8 $$

Solving step by step:

1. Square the term:

   $$ \left(-\frac{5}{8}\right)^2 = \frac{25}{64} $$

2. Multiply by 4:

   $$ 4 \times \frac{25}{64} = \frac{100}{64} = \frac{25}{16} $$

3. Multiply the linear term:

   $$ 5 \times -\frac{5}{8} = -\frac{25}{8} = -\frac{50}{16} $$

4. Convert the constant term:

   $$ 8 = \frac{128}{16} $$

5. Sum everything:

   $$ f(x_{\text{min}}) = \frac{25}{16} - \frac{50}{16} + \frac{128}{16} $$

   $$ = \frac{103}{16} \approx 6.44 $$

### **Conclusion**
- The function has a **local minimum at** \( x = -\frac{5}{8} \).
- The **minimum value** of \( f(x) \) is:

  $$ \frac{103}{16} \approx 6.44 $$

- Since the parabola opens upwards, **\( f(x) \geq 6.44 \) for all \( x \)**.


```python


##Initiate a ranom guess for local minima

local_minima = np.random.choice(x)
print(f"Initial guess for local minima is {local_minima}")

#Learning rate 
learning_rate = 0.01
epochs = 100
grad_list = []
local_minima_list = []
for i in range(epochs):
    grad = f_derivative_1st(local_minima)
    local_minima = local_minima - learning_rate * grad
    grad_list.append(grad)
    local_minima_list.append(local_minima)
print(f"local minima is {local_minima}")
```

```python
plt.plot(np.array(grad_list, local_minima), marker='o', color='g')
plt.grid()
```

If \( f''(x) > 0 \) at a critical point, the function has a **local minimum** at that point.  

If \( f''(x) < 0 \) at a critical point, the function has a **local maximum** at that point.  

If \( f''(x) = 0 \), the test is **inconclusive** (it might be an **inflection point**).
